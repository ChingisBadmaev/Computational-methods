### 2. y = 150 * sin(x / 15) + 7

```python
Image(filename = "test2.png", width = 600)
```

```python
def f_2(x):
    return 150 * math.sin(x / 15) + 7 
```

```python
def gradient_2(x):
    return 10 * math.cos(x / 15)
```

### Градиентный спуск

```python
def gradient_descent(x_old, eps):
    k = 3
    x_new = x_old - gradient_2(x_old)
    while abs(x_new - x_old) > eps:
        x_old = x_new
        x_new = x_old - gradient_2(x_old) / k
        k += 1
    answer = [x_new, f_2(x_new), k]
    return answer
```

# Вывод данных

```python
def pr(x, epsilon):
    print('x = ', gradient_descent(x, epsilon)[0])
    print('y = ', gradient_descent(x, epsilon)[1])
    print('k = ', gradient_descent(x, epsilon)[2])
```

## 1. y = x ** 2

```python
pr(500, 10e-5)
```
